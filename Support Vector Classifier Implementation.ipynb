## SUPPORT VECTOR CLASSIFIER IMPLEMENTATION

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes= 2, n_clusters_per_class=2,n_redundant=0)
```

```python
X   # independent variables, we have two features can convert to dataframe
```

```python
y  # dependent variable
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue=y) 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(X_train, y_train)
```

```python
svc.coef_  # coefficients of the linear model
```

```python
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

### HYPERPARAMETER TUNING WITH SVC

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear'],
    'gamma': ['scale', 'auto']
}
```

```python
grid = GridSearchCV(SVC(), param_grid = param_grid, refit=True, verbose=3, cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_  # best parameters found by grid search
```

```python
## Prediction
y_pred_grid = grid.predict(X_test)
print(classification_report(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
print(accuracy_score(y_test, y_pred_grid))
```
